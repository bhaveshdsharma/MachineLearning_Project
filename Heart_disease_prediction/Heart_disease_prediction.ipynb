## Using some hyperparameter tuning (Randomized Search CV) that will try to tune the hyperparamets for RandomForestClasiffier. 

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier = RandomForestClassifier(n_jobs=-1)
rf_hyp={'max_depth':[3,4,5,6,8,None],
        
           'n_estimators':[80,100,150,200,300],
        
            'criterion': ['entropy','gini']
       }
```

```python
RSC=RandomizedSearchCV(classifier,param_distributions=rf_hyp,n_iter=40,cv=10)
```

```python
RSC.fit(X,Y)
```

```python
RSC.best_score_
```

```python
RSC.best_params_
```

```python
classifier2 = RandomForestClassifier(n_jobs=-1,n_estimators= 150, max_depth= 5, criterion= 'entropy')
```

```python
from sklearn.model_selection import cross_val_score
CVS= cross_val_score(classifier2,X,Y,cv=20)
```

```python
print(CVS)
```

```python
CVS.mean()
```

## Splitting the data into training and testing data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=.20,random_state=0)
```

```python
classifier2.fit(X_train,Y_train)
```

```python
Y_pred=classifier2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy=accuracy_score(Y_test,Y_pred)
cfm=confusion_matrix(Y_test,Y_pred)
print(accuracy)
print(cfm)
```

### Plotting boxplots to see  any outliers in our data and handaling it using IQR technique
### (considering data betwen 25th and 75th percentile as non outlier)

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20,15))
ax = ax.flatten()
index = 0
for i in df.columns:
    sns.boxplot(y=i, data=df, ax=ax[index])
    index +=1
plt.tight_layout(pad=0.4)    
```

```python
# We can use Min max normalization to reduce the impact of outlier
```

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20,10))
ax = ax.flatten()
index = 0
for col,value in df.items():
    sns.distplot(value, ax=ax[index],kde=False)
    index +=1  
```

## Splitting the data into Dependent and Independent variable

```python
X = df1.drop(['Target','RAD'], axis = 1)
y = df1['Target']
```

```python
X.head()
```

```python
y.head()
```

### Training the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
X_trian,X_test,y_trian,y_test = train_test_split(X,y,test_size=.20,random_state=30)
```

```python
LR.fit(X_trian,y_trian)
```

```python
pred=LR.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE       :', mean_squared_error(y_test,pred))
print('R-Squared :', r2_score(y_test,pred))
```
